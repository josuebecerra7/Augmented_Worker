```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
import os

for path in [config_file, frozen_model]:
    print(os.path.exists(path))
```

```python
model = cv2.dnn_DetectionModel(frozen_model, config_file)
```

```python
classLabels = []
file_name = 'coco.names'
with open(file_name,'rt') as ftp:
    classLabels = ftp.read().rstrip('\n').split('\n')
```

```python
print(len(classLabels))
```

```python
# Resize the images according to the configuration model (320x320)pixels
model.setInputSize(320,320)
model.setInputScale(1.0/127.5) # 255/2 = 127.5
model.setInputMean((127.5,127.5,127.5)) # mobilenet takes [-1,1]
model.setInputSwapRB(True)
```

# read an image

```python
img = cv2.imread('test_image.jpg')
```

```python
plt.imshow(img) # brg colour format
```

```python
# Print with real colours
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# Set the accuracy thereshold
ClassIndex, confidence, bbox = model.detect(img, confThreshold=0.4)
```

```python
print(classLabels[ClassIndex.flatten()[0]])
```

```python
font_scale = 1.5
font = cv2.FORMATTER_FMT_PYTHON
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    cv2.rectangle(img, boxes, (255,0,0), 2)
    cv2.putText(img,classLabels[ClassInd-1],(boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(15,200,10), thickness=3 )
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# Read Image
img = cv2.imread('test_2.jpeg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# Read Image
img = cv2.imread('test_2.jpeg')
# Set the accuracy thereshold
ClassIndex, confidence, bbox = model.detect(img, confThreshold=0.55)
# Print tags
font_scale = 1.5
font = cv2.FORMATTER_FMT_PYTHON
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    cv2.rectangle(img, boxes, (255,0,0), 2)
    cv2.putText(img,classLabels[ClassInd-1],(boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(15,200,10), thickness=3)

# Show image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
